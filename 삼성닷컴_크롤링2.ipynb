```python
dfs = pd.DataFrame()

for url in urls:
    request = requests.get(url)
    soup = BeautifulSoup(request.text)
    divs = soup.find('div', {'class': 'toggle-wrap'})
    data = {}


    for div in divs:
        if isinstance(div, NavigableString):
            continue
        elif isinstance(div, Tag):
            strong = div.findAll('strong')

            for strong_tag in strong:
                key = strong_tag.text.strip()
                p = strong_tag.find_next('p')
                value = p.get_text().strip()

                if key in data:
                    data[key].append(value)
                else:
                    data[key] = [value]

    df_s2 = pd.DataFrame(data)
    dfs = pd.concat([dfs, df_s2], ignore_index=True)

pd.set_option('display.max_columns', None)
dfs
```
