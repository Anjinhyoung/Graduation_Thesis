# TB를 GB로 통일


```python
def convert_storage_to_gb(value):

    value_str = str(value).strip()

    if "TB" in value_str:
        return int(value_str.replace("TB", "").strip()) * 1000

    if "GB" in value_str:
        return int(value_str.replace("GB", "").strip())

    return int(value_str)

df['스토리지(저장 용량) (GB,TB)'] = df['스토리지(저장 용량) (GB,TB)'].apply(convert_storage_to_gb)
```

# 'CPU 속도'열에서 가장 높은 단위를 가진 숫자만 추출하기

```python
def extract_max_cpu_speed(value):
    # 'GHz'를 제거하고 쉼표로 분리하기
    speeds = [float(speed.replace('GHz', '').strip()) for speed in str(value).split(',')]

    # 가장 높은 단위 추출
    return max(speeds)

df['CPU 속도'] = df['CPU 속도'].apply(extract_max_cpu_speed)
```

# 전면 카메라 - 화소(MP)열 전처리
## 1. '+'로 분리하기
## 2. 숫자가 두 개 사이에 '+'가 있을 경우 더하기
## 3. 숫자 뒤에 '+'만 있는 경우 그대로 두기

```python
def process_plus(value):
    # '+'로 분리하기
    parts = [part.strip() for part in str(value).split('+')]

    # 숫자가 두 개 있을 경우 더하기
    if len(parts) == 2 and parts[1] != "":
        return str(float(parts[0]) + float(parts[1]))

    # 숫자 뒤에 '+'만 있는 경우 그대로 두기
    elif len(parts) == 2 and parts[1] == "":
        return parts[0]

    return value

df['전면 카메라 - 화소(MP)'] = df['전면 카메라 - 화소(MP)'].apply(process_plus)
```

# '해상도(픽셀 수)'열을 숫자형 데이터 타입으로 변환하기

```python
def calculate_resolution(resolution):
    # 대괄호([])와 그 안의 내용을 찾아 계산하고 대괄호를 제거
    if '[' in resolution and ']' in resolution:
        start_index = resolution.find('[')
        end_index = resolution.find(']')
        expression = resolution[start_index+1:end_index]
        evaluated_expression = eval(expression)
        resolution = resolution.replace(f'[{expression}]', str(evaluated_expression))

    # 'x'를 기준으로 문자열을 분할하고, 각 부분을 정수로 변환하여 곱하기
    parts = resolution.split('x')
    if len(parts) == 2:
        return int(parts[0].strip()) * int(parts[1].strip())

    return resolution

# 함수 적용
df['해상도(픽셀 수)'] = df['해상도(픽셀 수)'].apply(calculate_resolution)
```

```python
def classify_series(model_name):
    # 모델 이름이 'Galaxy'로 시작하고, 'S', 'A', 'Z' 중 하나로 이어지는 경우
    if model_name.startswith('Galaxy S'):
        return '갤럭시 S시리즈'
    elif model_name.startswith('Galaxy A'):
        return '갤럭시 A시리즈'
    elif model_name.startswith('Galaxy Z'):
        return '갤럭시 Z시리즈'
    else:
        return '갤럭시 기타 시리즈'

# '기종명(모델명)' 열의 각 값에 대해 classify_series 함수를 적용하여 '분류' 열을 생성합니다.
df['분류'] = df['기종명(모델명)'].apply(classify_series)
```

# K-means 성능의 향상을 위한 차원 축소

```python
from sklearn.decomposition import PCA
```

```python
# 선택한 열을 제외하고 데이터 준비
X = df.drop(columns=['기종명(모델명)', '출고가', '색상', '배터리 용량(mAh, Typical)', '최대 통화 시간 (시간)',
                    '인터넷 사용 시간(Wi-Fi)', '동영상 재생 시간', '오디오 재생 시간', '무게(g)',
                    '크기(Main Display)', '해상도(픽셀 수)', '전면 카메라 - 화소(MP)', 'CPU 속도', '분류'])

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans 클러스터링
kmeans = KMeans(n_clusters=61, random_state=0).fit(X_scaled)

# PCA로 데이터를 2차원으로 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 8))

# 각 클러스터에 대해 산점도 그리기
for cluster_num in range(4):
    members = X_pca[kmeans.labels_ == cluster_num]
    plt.scatter(members[:, 0], members[:, 1], label=f'Cluster {cluster_num}')

# 클러스터 중심 표시
cluster_centers_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(cluster_centers_pca[:, 0], cluster_centers_pca[:, 1], s=200, c='black', marker='X', label='Cluster Centers')

plt.title('KMeans Clustering with PCA')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')

plt.legend()

plt.show()
```

```python
# 클러스터 중심값 확인
cluster_centers = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=X.columns)
print(cluster_centers)
```

# DBSCAN 사용하기(전처리 과정 포함)

```python
import numpy as np

X = df.drop(columns=['기종명(모델명)', '출고가', '색상', '배터리 용량(mAh, Typical)', '최대 통화 시간 (시간)',
                    '인터넷 사용 시간(Wi-Fi)', '동영상 재생 시간', '오디오 재생 시간', '무게(g)',
                    '크기(Main Display)', '해상도(픽셀 수)', '전면 카메라 - 화소(MP)', 'CPU 속도', '분류'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

best_score = -1
best_eps = None
best_min_samples = None

# 여러 eps와 min_samples 값에 대해 실험
for eps in np.arange(0.1, 3, 0.1):  # eps 값을 0.1부터 2.9까지 0.1씩 증가
    for min_samples in range(2, 15):  # min_samples 값을 2부터 14까지 1씩 증가
        dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(X_scaled)
        labels = dbscan.labels_

        # 노이즈가 아닌 데이터 포인트에 대한 실루엣 스코어 계산
        if len(set(labels)) > 1:
            score = silhouette_score(X_scaled, labels)
            if score > best_score:
                best_score = score
                best_eps = eps
                best_min_samples = min_samples

# 최적의 파라미터 출력
print(f"Best Silhouette Score: {best_score}")
print(f"Best eps: {best_eps}")
print(f"Best min_samples: {best_min_samples}")
```

```python
X = df.drop(columns=['기종명(모델명)', '출고가', '색상', '배터리 용량(mAh, Typical)', '최대 통화 시간 (시간)',
                    '인터넷 사용 시간(Wi-Fi)', '동영상 재생 시간', '오디오 재생 시간', '무게(g)',
                    '크기(Main Display)', '해상도(픽셀 수)', '전면 카메라 - 화소(MP)', 'CPU 속도', '분류'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# DBSCAN 클러스터링
dbscan = DBSCAN(eps=1.5, min_samples=2).fit(X_scaled)

# PCA로 데이터를 2차원으로 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 8))

# 클러스터 번호 가져오기 (노이즈는 -1로 표시됨)
clusters = set(dbscan.labels_)

# 각 클러스터와 노이즈에 대해 산점도 그리기
for cluster_num in clusters:
    members = X_pca[dbscan.labels_ == cluster_num]
    if cluster_num == -1:
        label = 'Noise'
    else:
        label = f'Cluster {cluster_num}'
    plt.scatter(members[:, 0], members[:, 1], label=label)

plt.title('DBSCAN Clustering with PCA')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')

plt.legend()

plt.show()
```

# 계층적 클러스터링 사용하기(전처리 과정 포함)  

```python
from scipy.cluster.hierarchy import linkage, fcluster

X = df.drop(columns=['기종명(모델명)', '출고가', '색상', '배터리 용량(mAh, Typical)', '최대 통화 시간 (시간)',
                    '인터넷 사용 시간(Wi-Fi)', '동영상 재생 시간', '오디오 재생 시간', '무게(g)',
                    '크기(Main Display)', '해상도(픽셀 수)', '전면 카메라 - 화소(MP)', 'CPU 속도', '분류'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 계층적 클러스터링 수행
Z = linkage(X_scaled, method='ward')

# 클러스터 할당
max_d = 7  # 클러스터를 분할하는 최대 거리를 설정
clusters = fcluster(Z, max_d, criterion='distance')

# PCA로 차원 축소
pca = PCA(n_components=2).fit(X_scaled)
X_pca = pca.transform(X_scaled)

plt.figure(figsize=(10, 8))
for i in set(clusters):
    plt.scatter(X_pca[clusters == i, 0], X_pca[clusters == i, 1], label=f'Cluster {i}')
plt.legend()
plt.title('Hierarchical Clustering with PCA')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

# 계층적 클러스터링 실루엣 스코어 계산

```python
from sklearn.metrics import silhouette_score

sil_score_hierarchical = silhouette_score(X_scaled, clusters, metric='euclidean')

print(f"Silhouette Score (Hierarchical Clustering): {sil_score_hierarchical}")
```
