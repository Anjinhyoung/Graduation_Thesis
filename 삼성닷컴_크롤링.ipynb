```python
# 크롤링에 필요한 라이브러리와 모듈을 import하기

import pandas as pd
# pandas는 데이터 프레임을 만들기 위해서
import requests
from bs4 import BeautifulSoup,NavigableString, Tag
# requests,BeautifulSoup,NavigableString, Tag는 삼성 닷컴의 내용을 가져오고 파싱하기 위해서
```

```python
site = 'https://www.samsung.com/sec/support/model/SM-G900SZKESKO/'
# site 변수에 웹사이트의 URL을 저장
request = requests.get(site)
# requests.get(site)를 통해 해당 사이트의 내용을 가져오고, 그 결과를 request 변수에 저장
print(request)
# 200이 출력되면 잘 가져왔다고 해석하면 된다.
```

```python
soup = BeautifulSoup(request.text)
# print(soup)
# HTML을 파싱하고 그 결과를 soup 변수에 저장
```

```python
divs = soup.find('div', {'class': 'toggle-wrap'})
# soup.find를 사용하여 클래스 이름이 'toggle-wrap'인 'div' 태그를 찾아 divs 변수에 저장
print(divs)
```

```python
data = {}
# 추출한 데이터를 저장할 빈 딕셔너리를 생성
```

```python
for div in divs:
    if isinstance(div, NavigableString):
        continue
    elif isinstance(div, Tag):
        strong = div.findAll('strong')
    # 'divs' 리스트에서 각각의 'div' 태그에 대해 반복하고 div 안에 'strong'태그를 찾는다.

        for strong_tag in strong:
            key = strong_tag.text.strip()
            p = strong_tag.find_next('p')
            value = p.get_text().strip()
            # 각 'strong' 태그에 대해 태그의 텍스트를 key에 저장하고 'strong' 태그 다음의 'p' 태그의 텍스트를 value에 저장
            if key in data:
                data[key].append(value)
            else:
                data[key] = [value]
            # key가 data 딕셔너리의 키로 존재하면 value에 해당 키의 값(리스트)에 추가 없으면 새로운 키-값 쌍을 추가.
```

```python
pd.set_option('display.max_columns', None)
# 데이터 프레임 열 끝까지 보기
```

```python
df = pd.DataFrame(data)
df
```
